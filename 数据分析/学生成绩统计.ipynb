```python
# 创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。
# 计算平均分、最高分、最低分，并找出高于平均分的学生人数。

# 生成随机成绩
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
values= np.random.randint(50, 101, 10)
indexes = []
for i in range(1, 11):
    indexes.append('学生'+str(i))
series = pd.Series(values, indexes)
series
```

```python
print('平均分：', series.mean())
print('最高分：', series.max())
print('最低分：', series.min())
print('高于平均分的学生人数：', series[series > series.mean()].count())
```
