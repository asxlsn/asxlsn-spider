```python
'''
股票价格分析
给定某股票连续10个交易日的收盘价Series:
计算每日收益率(当日收盘价/前日收盘价-1)
找出收益率最高和最低的日期
计算波动率(收益率的标准差)
'''
import pandas as pd
import numpy as np
```

```python
prices = pd.Series([102.3,103.5,105.1,104.8,106.2,107.0,106.5,108.1,109.3,110.2],index=pd.date_range('2025-07-01',periods=10))
prices
```

```python
# 收益率
a = prices.pct_change()
```

```python
# 收益率最高的日期
print(a.nlargest(1))
# 收益率最低的日期
print(a.nsmallest(1))
```

```python
# 波动率
print(a.std())
```
