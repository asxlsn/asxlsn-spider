```python
'''
给定某城市一周每天的最高温度Series，完成以下任务
找出温度超过30度的天数
计算平均温度
将温度从高到低排序
找出温度变化最大的两天
'''

import pandas as pd
import numpy as np
```

```python
temperatures =pd.Series([28, 31,29, 32, 30,27, 33],index=['周一','周二','周三','周四','周五','周六','周日'])
temperatures
```

```python
num = temperatures[temperatures>30].count()
print('温度超过30度的天数为：',num)
```

```python
print('平均温度为：',temperatures.mean())
t2 = temperatures.sort_values(ascending=False)
print('温度从高到低排序为：',t2)
t3 = temperatures.diff().abs()
print('温度变化最大的两天为：',t3.nlargest(2))
```
